```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
teams_df = pd.read_csv('Team.csv')
team_attributes_df = pd.read_csv('Team_Attributes.csv')

merged_df = pd.merge(teams_df, team_attributes_df, on='team_api_id', how='inner')

#merged_df = merged_df[merged_df['overall_rating'] >= 82]

merged_df
```

```python
merged_df.drop_duplicates(subset=['team_long_name'], inplace=True)
# merged_df = merged_df.dropna()
name_df = merged_df.set_index('team_long_name')

name_df
```

```python
# drop irrelevant columns
df = name_df.drop(columns=['id_y', 'team_fifa_api_id_y', 'id_x', 'team_api_id', 'team_fifa_api_id_x', 'date', 'team_short_name'])

# drop non-quantitative attributes and buildUpPlayDribbling since it's almost fully null
df = df.drop(columns=['buildUpPlayDribbling', 'buildUpPlaySpeedClass', 'buildUpPlayDribblingClass', 'buildUpPlayPassingClass', 'buildUpPlayPositioningClass', 'chanceCreationPassingClass', 'chanceCreationCrossingClass', 'chanceCreationShootingClass', 'chanceCreationPositioningClass', 'defencePressureClass', 'defenceAggressionClass', 'defenceTeamWidthClass', 'defenceDefenderLineClass'])

df
```

```python
k = 5 # We pick k for illustrative purposes only. You would need a more principled approach here.
```

```python
cluster_model = KMeans(n_clusters=k)
```

```python
cluster_model.fit(df)
```

```python
cluster_labels = cluster_model.predict(df)
team_cluster_df = pd.DataFrame(cluster_labels, index=df.index, columns=["cluster"])
```

```python
team_cluster_df["cluster"].value_counts()
```

```python
for cluster,teams in team_cluster_df.groupby("cluster"):
    print("Cluster:", cluster, "Size:", teams.shape[0])
    
    top_attributes = df.loc[teams.index].sum().sort_values(ascending=False) / teams.shape[0]
    print("\t", "Attributes (High->Low):")
    for this_attribute,rating in top_attributes.items():
        print("\t\t", this_attribute, "[%0.4f]" % rating)
    
    print("\t", "Sample of Teams:")
    for team_id in teams.sample(10).index:
        print("\t\t", team_id)

    print("\n")
```
